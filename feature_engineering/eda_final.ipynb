**Data steel**

```python
data_steel[[
    'active_power',
    'reactive_power'
]] = data_steel[[
    'active_power',
    'reactive_power'
]].astype("float64")

data_steel['arc_heating_start'] = pd.to_datetime(data_steel['arc_heating_start'])
data_steel['arc_heating_end'] = pd.to_datetime(data_steel['arc_heating_end'])
```

```python
temp = np.datetime64(data_steel['arc_heating_end'][1]) - np.datetime64(data_steel['arc_heating_start'][1]) 
print(temp.astype('timedelta64[m]'))
print(temp / np.timedelta64(1, 'm'))
```

```python
# Calculate minutes
data_steel['minutes_process'] = data_steel.apply(lambda x: fa.calculate_time(x, 'arc_heating_start', 'arc_heating_end'), axis=1)
```

```python
# Calculate month
data_steel['months'] = data_steel['arc_heating_start'].dt.month 
```

```python
function_dictionary = {
    'active_power': ['sum'],
    'reactive_power': ['sum'],
    'minutes_process': ['sum']

}

new_val_col = [
    'active_power_sum',
    'reactive_power_sum',
    'minutes_process_sum'
]


data_group = data_steel.groupby([
    'key',
    'months'
]).aggregate(
    function_dictionary
)

data_group.columns = new_val_col
data_group = data_group.reset_index()
```
